# 使用Face-Recognition套件进行脸部侦测

## 载入相关套件

```python
# 安装套件： pip install face-recognition
# 载入相关套件
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import face_recognition
```

## 载入并显示图档

```python
# 载入图档
image_file = "./images_face/classmates.jpg"
image = plt.imread(image_file)

# 显示图像
plt.imshow(image)
plt.axis('off')
plt.show()
```

## 呼叫 face_locations 函数侦测脸部

```python
# 侦测脸部
faces = face_recognition.face_locations(image)
```

## 脸部加框，显示图像

```python
# 脸部加框
ax = plt.gca()
for result in faces:
    # 取得框的座标
    y1, x1, y2, x2 = result
    width, height = x2 - x1, y2 - y1
    # 加红色框
    rect = Rectangle((x1, y1), width, height, fill=False, color='red')
    ax.add_patch(rect)
        
# 显示图像
plt.imshow(image)
plt.axis('off')
plt.show()
```

## 侦测脸部特征点并显示

```python
# 侦测脸部特征点并显示
from PIL import Image, ImageDraw

# 载入图档
image = face_recognition.load_image_file(image_file)

# 转为 Pillow 图像格式
pil_image = Image.fromarray(image)

# 取得图像绘图物件
d = ImageDraw.Draw(pil_image)

# 侦测脸部特征点
face_landmarks_list = face_recognition.face_landmarks(image)

for face_landmarks in face_landmarks_list:
    # 显示五官特征点
    for facial_feature in face_landmarks.keys():
        print(f"{facial_feature} 特征点: {face_landmarks[facial_feature]}\n")

    # 绘制特征点
    for facial_feature in face_landmarks.keys():
        d.line(face_landmarks[facial_feature], width=5, fill='green')
    
# 显示图像
plt.imshow(pil_image)
plt.axis('off')
plt.show()
```
